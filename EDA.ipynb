```python
def get_data(path):
    train_data = []
    with open(path, encoding='utf-8') as f:
        words = []
        tags = []
        for line in f:
            line = line.strip()
            if not line:
                train_data.append([words, tags])
                words = []
                tags = []
            else:
                columns = line.split()
                words.append(columns[0])
                tags.append(columns[-1])
    return train_data

def get_instances_by_tag(dataset, tag):
    '''
        :param tag: ví dụ như JOB, ORGANIZATION, ....
        :return:
    '''
    # print(dataset)
    instances = []  # ví dụ job thì sẽ là: [[giáo viên], [công nhân], [y_tá, điều_dưỡng],...]
    for ex in dataset:
        tags = ex[1]
        for i in range(len(tags)):
            if tags[i] == f'B-{tag}':
                j = i + 1
                instance = [ex[0][i]]
                if j < len(tags):
                    while tags[j] == f'I-{tag}':
                        instance.append(ex[0][j])
                        j = j + 1
                        if j >= len(tags):
                            break

                # if instance not in instances:
                instances.append(' '.join(instance))

    return instances
```
